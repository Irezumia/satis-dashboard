```python
import numpy as np
import pandas as pd

np.random.seed(42)

# Tarih aralığı (son 9 ay)
dates = pd.date_range("2024-12-01", "2025-08-01", freq="D")

# Kategori ve ürünler
urunler = {
    "İçecek": ["Kahve", "Çay", "Limonata", "Soğuk Kahve"],
    "Atıştırmalık": ["Sandviç", "Kurabiye", "Muffin", "Tost"],
    "Yemek": ["Burger", "Pizza", "Salata", "Makarna"],
    "Tatlı": ["Cheesecake", "Tiramisu", "Dondurma"]
}

# Ürün bazlı temel fiyatlar (ortalama)
fiyat_tablosu = {
    "Kahve": 45, "Çay": 20, "Limonata": 35, "Soğuk Kahve": 55,
    "Sandviç": 70, "Kurabiye": 30, "Muffin": 38, "Tost": 60,
    "Burger": 140, "Pizza": 180, "Salata": 120, "Makarna": 130,
    "Cheesecake": 95, "Tiramisu": 110, "Dondurma": 60
}

kanallar = ["Mağaza", "Online"]
sehirler = ["İstanbul", "Ankara", "İzmir", "Bursa", "Antalya"]

kayıt_sayısı = 6000
tarih = np.random.choice(dates, kayıt_sayısı)

# Kategori-Ürün seçimi
kategoriler = np.random.choice(list(urunler.keys()), kayıt_sayısı)
urun_list = [np.random.choice(urunler[k]) for k in kategoriler]

# Fiyat: temel fiyat +/- %15 oynama
baz_fiyat = np.array([fiyat_tablosu[u] for u in urun_list])
fiyat = np.round(baz_fiyat * np.random.uniform(0.85, 1.15, kayıt_sayısı), 2)

# Adet (1-6 arası, içeceklerde daha küçük dağılım)
adet = np.where(np.isin(urun_list, urunler["İçecek"]),
                np.random.randint(1, 4, kayıt_sayısı),
                np.random.randint(1, 7, kayıt_sayısı))

kanal = np.random.choice(kanallar, kayıt_sayısı, p=[0.7, 0.3])
sehir = np.random.choice(sehirler, kayıt_sayısı, p=[0.45, 0.2, 0.2, 0.08, 0.07])

df = pd.DataFrame({
    "Tarih": pd.to_datetime(tarih),
    "Kategori": kategoriler,
    "Ürün": urun_list,
    "BirimFiyat": fiyat,
    "Adet": adet,
    "Kanal": kanal,
    "Şehir": sehir
})
df["Tutar"] = (df["BirimFiyat"] * df["Adet"]).round(2)

# Kaydet (notebook ile aynı klasöre)
df.to_csv("satislar.csv", index=False)
df.head()
```

```python
import pandas as pd

df = pd.read_csv("satislar.csv", parse_dates=["Tarih"])
print(df.shape)
df.head()
```

```python
# Tip kontrol
print(df.dtypes)

# Eksik değer var mı?
print("\nEksik değer sayıları:\n", df.isna().sum())

# Negatif/sıfır adet veya fiyat varsa temizle (olmaz ama kural koyalım)
df = df[(df["Adet"] > 0) & (df["BirimFiyat"] > 0)]

# Duplikeler
dup = df.duplicated().sum()
if dup > 0:
    df = df.drop_duplicates()
    print(f"{dup} duplike satır silindi.")

df.describe(include="all")
```

```python
toplam_gelir = df["Tutar"].sum()
siparis_sayisi = len(df)
ortalama_sepet = df["Tutar"].mean()

print(f"Toplam Gelir: ₺{toplam_gelir:,.0f}")
print(f"Sipariş Sayısı: {siparis_sayisi:,}")
print(f"Ortalama Sepet (AOV): ₺{ortalama_sepet:,.2f}")

top_urun = df.groupby("Ürün")["Tutar"].sum().sort_values(ascending=False).head(10)
top_kategori = df.groupby("Kategori")["Tutar"].sum().sort_values(ascending=False)

top_urun, top_kategori
```

```python
import matplotlib.pyplot as plt

# Aylık gelir
aylik = (df
         .set_index("Tarih")
         .resample("MS")["Tutar"]
         .sum())

plt.figure()
aylik.plot(marker="o")
plt.title("Aylık Toplam Gelir")
plt.xlabel("Ay")
plt.ylabel("₺")
plt.tight_layout()
plt.show()

# Kategori bazlı gelir
plt.figure()
top_kategori.plot(kind="bar")
plt.title("Kategori Bazlı Toplam Gelir")
plt.ylabel("₺")
plt.tight_layout()
plt.show()

# En çok ciro yapan 10 ürün
plt.figure()
top_urun.plot(kind="bar")
plt.title("En Çok Ciro Yapan 10 Ürün")

plt.ylabel("₺")
plt.tight_layout()
plt.show()
```

```python
with pd.ExcelWriter("rapor_satış.xlsx", engine="openpyxl") as writer:
    df.to_excel(writer, sheet_name="HamVeri", index=False)
    aylik.reset_index().to_excel(writer, sheet_name="AylıkGelir", index=False)
    top_urun.reset_index(name="ToplamTutar").to_excel(writer, sheet_name="TopUrunler", index=False)
    top_kategori.reset_index(name="ToplamTutar").to_excel(writer, sheet_name="TopKategoriler", index=False)

"rapor_satış.xlsx kaydedildi."
```
